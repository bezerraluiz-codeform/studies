```python
from langchain_text_splitters import MarkdownHeaderTextSplitter, RecursiveCharacterTextSplitter
```

```python
text_ocr_markdown = """
# CONTRATO DE ALUGUEL

## Cláusula de Pagamento
O pagamento deve ser feito até o dia 5.
Em caso de atraso, multa de 2%.

## Cláusula de Rescisão
A rescisão pode ser feita com aviso prévio de 30 dias.
Caso não haja aviso, será cobrada multa de um aluguel.
"""
```

```python
headers_to_split_on = [
    ("#", "Principal Header"),
    ("##", "Secondary Header"),
]
```

```python
markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
md_header_splits = markdown_splitter.split_text(text_ocr_markdown)
```

```python
print(f"--- Chunk Estrutural 1 ---")
print(f"Conteúdo: {md_header_splits[0].page_content}")
print(f"Metadados: {md_header_splits[0].metadata}")
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100, # Tamanho de caracteres por chunk
    chunk_overlap=20,
)
```

```python
final_chunks = text_splitter.split_documents(md_header_splits)
```

```python
print(f"--- Chunk Final Processado ---")
print(f"Conteúdo: {final_chunks[0].page_content}")
print(f"Metadados Ricos: {final_chunks[0].metadata}")
```
