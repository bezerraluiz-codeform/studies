```python
import numpy as np
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
print("Carregando o modelo...")
```

```python
# Configuração do modelo
model_name = "BAAI/bge-m3"
model_kwargs = {'device': 'cuda'}
encode_kwargs = {'normalize_embeddings': True}
```

```python
# Instanciando
bge_embeddings = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
print("Modelo carregado com sucesso!")
```

# Teste de Similiaridade Semântica

```python
# Frase 1: O que o usuário busca
query_text = "Tênis de corrida para maratona?"

# Documentos
documents = [
    "Tênis para maratona, confortável da Nike com amortecimento de alta performance",
    "Camiseta dry fit para treino",
    "Tênis com placa de carbono para maratona e longa distância",
]
```

```python
# Gera embeddings
query_vector = np.array(bge_embeddings.embed_query(query_text), dtype=np.float32)
# `embed_documents` espera uma lista de strings (um texto por documento)
document_vectors = np.array(bge_embeddings.embed_documents(documents), dtype=np.float32)

print(query_vector)
print(document_vectors)
```

```python
# Simulando cálculo de similaridade de cosseno realizado por um banco de dados
# Como normalizamos antes, basta fazer o produto escalar (dot product) entre os vetores
# score = float(np.dot(queryVector, documentVector))

score = document_vectors @ query_vector

print(score)
```

```python
best_idx = int(np.argmax(score))
best_document = documents[best_idx]
best_score = float(score[best_idx])
```

```python
print(f"Documento com maior score: {best_document}")
print(f"Maior score: {best_score:.4f}")
```

```python
# Interpretação
if best_score > 0.5:
    print("O documento é relevante para a query.")
else:
    print("O documento não é relevante para a query.")
```
